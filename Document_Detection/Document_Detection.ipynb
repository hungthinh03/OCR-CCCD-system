```python
import numpy as np
import torch

print("NumPy:", np.__version__)
print("Torch:", torch.__version__)
print("CUDA:", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0))
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/kaggle/working/dataset.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    name="final_document_detector"
)
```

```python
metrics = model.val()

print(metrics)
```

```python
import shutil

shutil.make_archive(
    "/kaggle/working/final_document_detector",
    'zip',
    "/kaggle/working/runs/detect/final_document_detector"
)
```

```python
metrics = model.val()
print(metrics.box.map)      # mAP50-95
print(metrics.box.map50)    # mAP@0.5
print(metrics.box.mp)       # Precision
print(metrics.box.mr)  
```
